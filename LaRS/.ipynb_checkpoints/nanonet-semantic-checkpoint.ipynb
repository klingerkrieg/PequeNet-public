```python
info_dataset_resolution=256
info_dataset_batch_size=8
train_loader, test_loader = get_LaRS_datasets(dataset_dir=get_dataset_path(), 
                                              resolution=info_dataset_resolution, 
                                              batch_size=info_dataset_batch_size)
```

```python
model_name   = 'SemanticNanoNet1'
path         = './SemanticNanoNet'
features_int = 8 
mlp_int      = 10
net_size     = 2
            
config = f'features_int={features_int}, mlp_int={mlp_int}, net_size={net_size}'
model = SemanticNanoNet1(in_channels=3, out_channels=3, 
                         features_int=features_int, 
                         mlp_int=mlp_int, 
                         net_size=net_size)

model_filename = f'{model_name}-fi{features_int}-mlpi{mlp_int}-net_size{net_size}-res{info_dataset_resolution}-epochs{epochs}.pth'
trainer = SemanticTrainer(num_classes=3,
                model_filename=model_filename, 
                model_dir=path, 
                info={'dataset_resolution':info_dataset_resolution,
                        'dataset_batch_size':info_dataset_batch_size,
                        'config':config,
                        'trainable_parameters':count_trainable_parameters(model),
                        'dataset_name':dataset_name,
                        },
                save_xlsx=True)
model = trainer.train(model, test_loader, train_loader, 
                        num_epochs=epochs, 
                        verbose=3, 
                        save_every=1, 
                        continue_from_last=True)
model.to(device)
#trainer.save_sample_output(test_loader,[0,1,2,3,4])

#Tempo médio por imagem: 11.241 ms
#Imagens por segundo (FPS): 88.96
```
